## 4 Metrics

In order to evaluate performance of a Machine Learning regression model I will be using Root Mean Squared Error (RMSE) and Mean Absolute Error (MAE). For both metrics, the units of the error score match the units of the target value that is being predicted which makes the model performance interpretation easier. The RMSE and the MAE can be used together to diagnose the variation in the errors in a set of forecasts. The RMSE will always be larger or equal to the MAE; the greater the difference between them, the greater the variance in the individual errors in the sample.

***Root Mean Squared Error (RMSE)***

The RMSE is a quadratic scoring rule that measures the average magnitude of the error. RMSE is the difference between forecast and corresponding observed values squared and then averaged over the sample. Finally, the square root of the average is taken. Since the errors are squared before they are averaged, the RMSE gives a relatively high weight to large errors. This means the RMSE is most useful when large errors are particularly undesirable.

***Mean Absolute Error (MAE)***

The MAE measures the average magnitude of the errors in a set of forecasts, without considering their direction. It measures accuracy for continuous variables. MAE is the average over the verification sample of the absolute values of the differences between forecast and the corresponding observation. The MAE is a linear score which means that all the individual differences are weighted equally in the average.

All electric fueled cars are model i3 but they show the exact same mpg value as hybrid cars which is strange since different fuels should provide different efficiencies. I will aggregate electric fueled i3 cars to i3 hybrid cars while leaving the "other" fuel type as a separate category.

## 6 Hypothesis Testing

### 6.1 Fuel type and price

Since diesel and petrol cars are the most common in the dataset, it would be interesting analyzing whether there is a difference in price and, in such case, evaluating whether the difference is statistically significant.

```python
# Displaying price boxplots and distributions over petrol and diesel fuel types
df_petr_dies = df[(df.fuelType == 'Petrol') | (df.fuelType == 'Diesel')]

fig, ax = plt.subplots(1, 2, figsize = (10,5))
for i in ['Diesel', 'Petrol']:
    df_petr_dies[df_petr_dies.fuelType == i]['price'].plot(kind = 'kde', ax = ax[1], label = i)
sns.boxplot(x = 'fuelType', y = 'price', data = df_petr_dies, ax = ax[0])
ax[1].set_xlabel('price')
plt.suptitle('Price distribution over petrol/diesel fuels', fontsize = 'xx-large')
plt.legend()
plt.tight_layout();
```

Petrol and diesel fueled cars show different distributions. Let's see whether we can confirm it with hypothesis testing.

***Ttest***

A t-test is a statistical test that is used to compare the means of two groups. It is often used in hypothesis testing to determine whether a process or treatment actually has an effect on the population of interest, or whether two groups are different from one another. 

The t-test assumes that:
- data are independent
- data are approximately normally distributed.
- data have a similar amount of variance within each group being compared (homogeneity of variance)

Petrol fueled cars prices and Diesel fueled cars prices are independent and their distributions, as shown above, can be considered normal. Let's explore their variances:

```python
# Generating two groups and calculating their variances
petrol_price = df_petr_dies[df_petr_dies.fuelType == "Petrol"]["price"]
diesel_price = df_petr_dies[df_petr_dies.fuelType == "Diesel"]["price"]
print('Petrol fueled cars price variance: {:.2f}'.format(petrol_price.var()))
print('Diesel fueled cars price variance: {:.2f}'.format(diesel_price .var()))
```

Variances in the two groups are on the same scale, hence all the assumptions for running a ttest are met. Before running the test the null hypothesis and the alternative hypothesis need to be decided.
The null hypothesis is lack of difference in price distribution between petrol fueled and diesel fueled cars, whereas alternative hypothesis states such difference.

```python
# Importing ttest from library and performing it
from scipy.stats import ttest_ind
print(f'p_value: {ttest_ind(petrol_price, diesel_price)[1]}')
```

The two distributions seem to be different given the low p value or, in other words, the likelihood that the difference is due to chance is very low.

### 6.2 Transmission type and price

I will now analyze the differences on price in relation to the type of transmission which has only 3 categories.

```python
# Printing transmission relative frequencies
df.transmission.value_counts(normalize = True)
```

```python
# Displaying price boxplots and distributions over type of transmission
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
for i in ['Automatic', 'Manual', 'Semi-Auto']:
    df[df.transmission == i]['price'].plot(kind = 'kde', ax = ax[1], label = i)
sns.boxplot(x= 'transmission', y = 'price', data = df, ax = ax[0])
ax[1].set_xlabel('price')
plt.legend()
plt.suptitle('Price distribution over type of transmission', fontsize = 'xx-large')
plt.tight_layout()
```

The price seems to be different in relation to the type of transmission too. I will use hypothesis testing to exclude or confirm that the difference is due to chance.

***One Way ANOVA***

ANOVA, which stands for Analysis of Variance, is a statistical test used to analyze the difference between the means of more than two groups. ANOVA tells if the dependent variable changes according to the level of the independent variable. ANOVA assumptions are the same ones for ttest seen above. In this case too data will be considered independant and normally distributed. Below I will explore the variances of the three groups.

```python
# Generating three groups for the test and calculating variances
aut = df[df.transmission == 'Automatic']['price']
man = df[df.transmission == 'Manual']['price']
semi_aut = df[df.transmission == 'Semi-Auto']['price']

print('Automatic transmission price variance: {:,.2f}'.format(aut.var()))
print('Manual transmission price variance: {:,.2f}'.format(man.var()))
print('SemiAutomatic transmission price variance: {:,.2f}'.format(semi_aut.var()))
```

Unfortunately, although semiautomatic and automatic group variances are comparable, manual transmission variance is off scale and ANOVA test cannot be performed since not all the assumptions for the test are met.

### 6.3 Quantitative features exploration

Until now I focused the analysis on categorical features. Now I will explore quantitative features. Let's see whether I can spot interesting univariate correlation using Pearson's coefficient which is a measure of linear correlation between two sets of data.

```python
# Displaying a correlation heatmap
sns.heatmap(df.corr(), cmap = 'RdBu_r', annot = True, cbar = False);
```

We can consider an existing correlation between two variables when the absolute value of the Pearson's coefficient is higher than 0.3 (although there is no fixed threshold). The strongest correlation is the negative one between year and mileage which means the higher the mileage, the lower the year. This is quite expected since older cars tend to have higher mileage. Both features are also strongly correlated to the price feature, a phenomena known as multicollinearity. Since price and mileage are highly correlated, I will plot them with a third feature in order to spot any clusters.

```python
# Displaying a grid of scatterplots colored by a third feature
fig, axs = plt.subplots(2,2, figsize = (10, 8), sharey = True)
for ax, col in zip(axs.flatten(), ['year','transmission', 'fuelType']):
    sns.scatterplot(x = 'mileage', y = 'price', data = df, hue = col, ax = ax)
plt.tight_layout()
plt.suptitle('Price vs Mileage', fontsize = 'xx-large', y = 1.05)
fig.delaxes(axs[1][1])
```

As reported above older cars tend to have higher mileage and be cheaper. Segmenting by transmission types confirms some sort of stratification and higlights that cars with semiautomatic transmission tend to have lower mileage and higher price. Since the lower mileage is associated to higher price independently from the semiautomatic transmission type, this feature could be a confounding variable. Finally, no clusters are detectable segmenting data based on fuel type.

## 7 Machine Learning Pipeline Design, Fitting and Evaluation

The goal is builduing a model that predicts the price based on all the features in the dataset. The dataset is made up of both categorical and quantitative features. In order to perform machine learning often some preprocessing needs to be done since usually input data can only be numerical which means that categorical feautures need to be one hot encoded, for example. Exploring data further will allow to evaluate which preprocessing steps would be ideal.

```python
# Plotting quantitative datatypes distributions
df.select_dtypes(['float64', 'int64']).boxplot()
plt.title('Quantitative features boxplot');
```

Quantitative variables are on different scales thus scaling them could be a good idea since it can improve some types of algorithm performances. As seen in the graph above a lot of data could be considered outliers since the distributions are skewed. I will use robust scaling which allows feature scaling with statistics resistant to outliers i.e. removing the median and scaling to the interquartile range.

Let's make a recap of the categorical features which will need to be onehotencoded:

```python
# Displaying categorical features count plots
fig, axs = plt.subplots(1,3,figsize = (15,5))
for ax, col in zip(axs.flatten(), ['fuelType', 'transmission', 'model']):    
    sns.countplot(x = col, data = df, ax = ax)
    plt.xticks(rotation = 90)
plt.suptitle('Categorical features count plots', fontsize = 'xx-large')
plt.tight_layout()
```

As a note the model column has 24 categories which will produce 24 new columns once onehotencoded. Checking on the number of categories is always best practice before onehotencoding in order to evaluate how wide the dataframe will become.

```python
# Importing classes and functions I designed. For references, see func.py file
from func import RobustScalerCust, GetDummies, CustomEstimator, get_fitted_pipe, get_results

# Importing useful libraries
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
import catboost as cb
import lightgbm as lgb
```

I will proceed building a pipeline with the steps described above (scaling and onehotencoding). The final step of the pipeline will be an estimator. 
I will use different algorithms, both linear and non linear, along with hyperparameters selected by grid searching and performing 5 fold cross validation and compare their performances.

```python
# Implementing the pipeine
pipe = make_pipeline(RobustScalerCust(col_type = ['float64', 'int64']),
                     GetDummies(all_data = df, cols = ['model','fuelType','transmission']),
                     CustomEstimator())
```

The dataframe will be splitted in input features and target feature (X and y respectively) and in train and test set (80% and 20% of the data respectively)

```python
# Splitting the dataframe 
X = df.drop('price', axis = 1)
y = df.price
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

I will now implement the parameters to be used in grid searching by defining an estimator and its specific hyperparameters. Due to computational constraints I will limit them to only a few.

```python
# Defining param dictionaries for the param_grid attribute in GridSearch
params_lr = {'customestimator': [LinearRegression()]}

params_knn = {'customestimator': [KNeighborsRegressor()],
                                  'customestimator__n_neighbors': [i for i in range(3,8)],
                                  'customestimator__algorithm': ['auto', 'brute']}
          
params_xgb = {'customestimator': [xgb.XGBRegressor(n_estimators = 10000)]}
          
params_cb = {'customestimator': [cb.CatBoostRegressor(verbose = False)]}
          
params_lgb = {'customestimator': [lgb.LGBMRegressor()],
                                 'customestimator__n_estimators':[100, 500, 1000]}

params_gb = {'customestimator': [GradientBoostingRegressor(n_estimators=500)]}
```

```python
# Defining a dict containing the defined pipeline and the parameters
pipe_dict = {'pipe_lr': [pipe, params_lr],
             'pipe_knn': [pipe, params_knn],
             'pipe_xgb': [pipe, params_xgb],
             'pipe_cb': [pipe, params_cb],
             'pipe_lgb': [pipe, params_lgb],
             'pipe_gb': [pipe, params_gb]}
```

I will proceed fitting all the defined pipelines and parameters to GridSearch which will hold best estimators and best parameters for each and all algorithms based on cross validation in the train set using a helper function.

```python
fitted_pipe = get_fitted_pipe(pipe_dict, X_train, X_test, y_train, 'neg_root_mean_squared_error')
```

After fitting all the pipelines, another helper function outputs a dataframe with RMSE and MAE results obtained predicting the unseen test set with algorithms and parameters that performed the best in the train set. Performances for each algorithm can then be plotted.

```python
# Plotting MAE and RMSE results
results_df = get_results(fitted_pipe, X_test, y_test).sort_values('rmse', ascending = False)
results_reshape = results_df.reset_index().melt(id_vars = 'index', var_name = 'metric', value_name = 'error')
sns.catplot(data = results_reshape, x = 'index', y = 'error', hue = 'metric', kind = 'bar')
plt.title('Algorithms performance in test set')
plt.xlabel(None);
```

Algorithms able to grasp non linearity perform better. CatBoost obtains the best scores according to both MAE and RMSE.

```python
# Getting exact error values for CatBoost
print('CatBoost has a RMSE of {:.2f} and a MAE of {:.2f}'.format(results_df.loc["pipe_cb", "rmse"], 
                                                                 results_df.loc["pipe_cb", "mae"]))
```

Let's compare errors in training and testing phases for this algorithm:

```python
# Comparing CatBoost train and test error 
print('CatBoost training RMSE: {:.2f}'.format(- fitted_pipe["pipe_cb"].best_score_))
print('CatBoost test RMSE: {:.2f}'.format(results_df.loc['pipe_cb', 'rmse']))
```

Training and test errors are very close and we can exclude overfitting. In order to properly evaluate the results, it is helpful to consider LinearRegression as an unskilled learner and setting it as a benchmark in relation to which comparing the best performing estimator.

```python
# Calculating percentage error difference between LinearRegression and CatBoost
lr_cb = results_df.loc[('pipe_lr', 'pipe_cb'),:].T
lr_cb['diff%'] = round((lr_cb.pipe_lr - lr_cb.pipe_cb)/lr_cb.pipe_lr*100, 2)

print('Mean error reduction using CatBoost: {} %'.format(lr_cb["diff%"].mean()))
```

Having collected a dictionary with the fitted pipelines, it is easy to access to pipeline steps to gather CatBoost feature importances:

```python
# Plotting CatBoost feature importances
feat_imp = pd.DataFrame(fitted_pipe['pipe_cb'].best_estimator_.steps[2][1].feature_importances_,
                        index = fitted_pipe['pipe_cb'].best_estimator_.steps[1][1].new_cols,
                        columns = ['feature_importance'])
feat_imp_subset = feat_imp[feat_imp.feature_importance > 1].sort_values(by = 'feature_importance', ascending = False)
sns.barplot(data = feat_imp_subset, x = 'feature_importance', y = feat_imp_subset.index, palette = 'winter_r')
plt.title('Feature importances from CatBoost');
```

The graph above shows feature importances greater than 1. Quantitative variables with the exclusion of tax feature concur the most to the model while categorical features seem to have a lower impact. We can say that cars years, engine size and miles per gallon are the strongest variable for predicting cars prices. Apparently, although the ttest confirmed price differences among different fuel types, this feature is not one of the most important one when building a model with CatBoost.
